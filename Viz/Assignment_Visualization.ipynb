```python
import pandas as pd
```

```python
data_chilla = pd.read_csv('Chilla_data2_for_plots.csv')
data_chilla.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x = "Your favorite programming language?", y = "Age (years)-Float/Int", data = data_chilla)
plt.show()
```

```python
sns.barplot(x = "Marital Status?", y = "Height in cm? Freelancer- (Float)", hue = "Are you Vaccinated?", data = data_chilla)
plt.show()
```

```python
from numpy import mean
sns.barplot(x = "How many hours you code a day? (int) e.g: 5,4,3", y = "What are you?", data = data_chilla, estimator = mean, saturation = 2)
plt.show()
```

```python
# import library
import seaborn as sns
import numpy as np
import pandas as pd
sns.set(style = "whitegrid")
sns.boxplot(x = "How many hours you code a day? (int) e.g: 5,4,3", y = "Age (years)-Float/Int", data = data_chilla, saturation = 2)
```

```python
# import library
import seaborn as sns
import numpy as np
import pandas as pd
sns.set(style = "whitegrid")
sns.boxplot(x = "Light kitni der band hti hy? int", y = "Where do you live?", hue = "Your favorite programming language?", data = data_chilla, saturation = 2, palette = "Set2", dodge = True, orient = "h")
```

```python
#draw a line plot
sns.lineplot(x = "Light kitni der band hti hy? int", y = "How many hours you code a day? (int) e.g: 5,4,3", data = data_chilla)
plt.show()
```
